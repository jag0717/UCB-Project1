```python
# dependencies
import csv
import os
import pandas as pd
from pandas import *
from IPython.display import display
import matplotlib
```

```python
# load csv file
# census file for population 2017 path_1
# census file for population 2007 path_2
# unemployment rate 2017 path_3
# cost of living indexes 2017 path_4

file_path_1 = os.path.join('nst-est2017-alldata.csv')
file_path_2 = os.path.join('nst-census00to09-2.csv')
file_path_3 = os.path.join('unemployment-2017.csv')
file_path_4 = os.path.join('cost-of-living-2017.csv')

# read the csv file
population_2017 = pd.read_csv(file_path_1, encoding = "ISO-8859-1")
population_2007 = pd.read_csv(file_path_2, header = 3, encoding = "ISO-8859-1")
unemployment_2017 = pd.read_csv(file_path_3, encoding = "ISO-8859-1")
cost_living_2017 = pd.read_csv(file_path_4, header = 1, encoding = "ISO-8859-1")
```

```python
# display files and rename columns
# Create new dataframe with only the columns of interest
population_2017 = population_2017 [["NAME","POPESTIMATE2017"]]
population_2017 = population_2017.rename(columns = {'NAME':'State','POPESTIMATE2017':'Population_2017'})
population_2017.dtypes
population_2017.head()
```

```python
# display files and rename columns
# Create new dataframe with only the columns of interest
population_2007 = population_2007 [["Unnamed: 0","2007"]]
population_2007 = population_2007.rename(columns = {'Unnamed: 0':'State','2007':'Population_2007'})

# always remember to check the types of columns for future calculation purposes 
population_2007["Population_2007"] = pd.to_numeric(population_2007['Population_2007'], errors='coerce')
population_2007.dtypes
population_2007
```

```python
# display files and rename columns
# Create new dataframe with only the columns of interest
unemployment_2017 = unemployment_2017.drop(['Rank'],axis = 1)
unemployment_2017 = unemployment_2017.rename(columns = {'November 2017(P)':'Unemployment_2017'})
unemployment_2017["Unemployment_2017"] = pd.to_numeric(unemployment_2017["Unemployment_2017"], errors='coerce')
unemployment_2017.head()
```

```python
# display files and rename columns
# Create new dataframe with only the columns of interest
cost_living_2017 = cost_living_2017[["State","Index"]].rename(columns = {'Index':'Cost_of_living_2017'})
cost_living_2017.head()
```

```python
# merge files based on states
merge_1 = pd.merge(population_2007,population_2017,how="outer",on="State")
merge_2 = pd.merge(merge_1,unemployment_2017,how="outer",on="State")
merge_all_census = pd.merge(merge_2,cost_living_2017,how="outer",on="State")
merge_all_census.head()
```

```python
census_all_data = merge_all_census.dropna(how='any').set_index('State')
census_all_data.head()
#census_all_data.dtypes
```

```python
# Print csv file
census_all_data.to_csv("Census_Consol_Data.csv")
```
